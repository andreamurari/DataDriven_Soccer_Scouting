```python
import time
import pandas as pd
import soccerdata as sd

# ==========================================
# 1. CONFIGURATION
# ==========================================
LEAGUES = [
    'ENG-Premier League',
    'ESP-La Liga',
    'ITA-Serie A',
    'GER-Bundesliga',
    'FRA-Ligue 1',
]
SEASONS = ['2024-2025', '2023-2024', '2022-2023']
STAT_TYPES = ['standard', 'shooting', 'passing', 'defense', 'possession']

print('--- START PROCESSING ---')
print(f'Leagues: {LEAGUES}')
print(f'Seasons: {SEASONS}')

# ==========================================
# 2. DOWNLOAD STATS (BY SEASON)
# ==========================================
print("\n[1/4] Downloading FBref tables...")

stats_by_type = {stat_type: [] for stat_type in STAT_TYPES}

for season in SEASONS:
    print(f"\n--- Season {season} ---")
    try:
        fbref = sd.FBref(leagues=LEAGUES, seasons=[season])
        for stat_type in STAT_TYPES:
            print(f"   - Downloading '{stat_type}'...")
            df_stat = fbref.read_player_season_stats(stat_type=stat_type)
            stats_by_type[stat_type].append(df_stat)
        time.sleep(1)
    except Exception as exc:
        print(f"   - ERROR in season {season}: {exc}")
        print('   - Continuing with next season...')
        continue

if not stats_by_type['standard']:
    raise RuntimeError('No season downloaded successfully.')

stats_std = pd.concat(stats_by_type['standard'])
stats_shoot = pd.concat(stats_by_type['shooting'])
stats_pass = pd.concat(stats_by_type['passing'])
stats_def = pd.concat(stats_by_type['defense'])
stats_poss = pd.concat(stats_by_type['possession'])

# ==========================================
# 3. DATAFRAME MERGE (SMART MERGE)
# ==========================================
print("\n[2/4] Merging tables...")

df_total = stats_std.copy()
other_dfs = [stats_shoot, stats_pass, stats_def, stats_poss]

for df_other in other_dfs:
    cols_to_add = df_other.columns.difference(df_total.columns)
    df_total = df_total.join(df_other[cols_to_add])

print(f"   - Tables merged. Current shape: {df_total.shape}")

# ==========================================
# 4. COLUMN HEADER CLEANUP
# ==========================================
print("\n[3/4] Cleaning column names...")

new_columns = []
for col in df_total.columns:
    if isinstance(col, tuple):
        clean_name = '_'.join(str(part) for part in col if part).strip()
    else:
        clean_name = str(col).strip()
    new_columns.append(clean_name)

df_total.columns = new_columns
df_total = df_total.reset_index()

# ==========================================
# 5. FILTERS (DATA QUALITY)
# ==========================================
print("\n[4/4] Applying filters...")

initial_count = len(df_total)

candidate_min_columns = ['Playing Time_Min', 'Min', 'Playing Time_MP']
minutes_col = next((col for col in candidate_min_columns if col in df_total.columns), None)

if minutes_col is not None:
    df_total = df_total[df_total[minutes_col] >= 900]
    print(f"   - Filtered players with < 900 minutes (column '{minutes_col}')")
else:
    print('   - WARNING: minutes column not found, filter skipped.')

if 'Pos' in df_total.columns:
    df_total = df_total[df_total['Pos'] != 'GK']
    print('   - Goalkeepers removed (GK)')

final_count = len(df_total)
print(f"   - Remaining players: {final_count} (from {initial_count})")
```

```python
import kagglehub

print('Downloading EAFC26 dataset from Kaggle...')
path = kagglehub.dataset_download('flynn28/eafc26-player-database')
print(f'Dataset path: {path}')
```

```python
import os
import re
import unicodedata
import pandas as pd

# ==========================================
# 0. PATH CONFIGURATION
# ==========================================
path_kaggle = path
file_fifa = os.path.join(path_kaggle, 'EAFC26-Men.csv')

# ==========================================
# 1. CLEANING FUNCTIONS
# ==========================================
MOJIBAKE_PATTERN = re.compile(r'Ã.|Â\s|â.|�')

def fix_mojibake(text):
    if pd.isna(text):
        return ''

    result = str(text)
    for _ in range(3):
        if not MOJIBAKE_PATTERN.search(result):
            break
        try:
            candidate = result.encode('latin1').decode('utf-8')
            if candidate == result or candidate.strip() == '':
                break
            result = candidate
        except (UnicodeEncodeError, UnicodeDecodeError):
            break

    return unicodedata.normalize('NFC', result)

def normalize_name(name):
    if pd.isna(name):
        return ''
    name = fix_mojibake(name).lower().strip()
    return ''.join(
        char for char in unicodedata.normalize('NFD', name)
        if unicodedata.category(char) != 'Mn'
    )

def extract_metric_value(value, unit):
    if pd.isna(value):
        return pd.NA

    text = str(value).lower().replace(',', '.').strip()
    match_with_unit = re.search(rf"(\d+(?:\.\d+)?)\s*{unit}", text)
    if match_with_unit:
        return float(match_with_unit.group(1))

    left_side = text.split('/')[0]
    match_number = re.search(r"(\d+(?:\.\d+)?)", left_side)
    if match_number:
        return float(match_number.group(1))

    return pd.NA

def detect_column(candidates, columns):
    for candidate in candidates:
        if candidate in columns:
            return candidate
    return None

# ==========================================
# 2. LOAD DATASETS
# ==========================================
print('Loading datasets...')
df_fbref = df_total.copy()
print('FBref dataset loaded successfully!')

try:
    df_fifa = pd.read_csv(file_fifa)
    print('EAFC26 dataset loaded successfully!')
except FileNotFoundError:
    raise FileNotFoundError(f'File not found: {file_fifa}')

# ==========================================
# 3. NAME PREPARATION
# ==========================================
print('\nNormalizing names for merge...')

fbref_name_col = 'player' if 'player' in df_fbref.columns else 'Player'
df_fbref[fbref_name_col] = df_fbref[fbref_name_col].apply(fix_mojibake)
df_fbref['nome_join'] = df_fbref[fbref_name_col].apply(normalize_name)

fifa_name_col = detect_column(['Name', 'FullName', 'short_name', 'name'], df_fifa.columns)
if fifa_name_col is None:
    fifa_name_col = df_fifa.columns[0]

df_fifa[fifa_name_col] = df_fifa[fifa_name_col].apply(fix_mojibake)
df_fifa['nome_join'] = df_fifa[fifa_name_col].apply(normalize_name)

# ==========================================
# 4. PHYSICAL COLUMNS
# ==========================================
height_col = detect_column(['Height'], df_fifa.columns)
weight_col = detect_column(['Weight'], df_fifa.columns)
foot_col = detect_column(['Preferred foot', 'Foot'], df_fifa.columns)
pos_col = detect_column(['Position'], df_fifa.columns)

print(f"EAFC columns -> Height: '{height_col}', Weight: '{weight_col}', Foot: '{foot_col}', Position: '{pos_col}'")

if height_col:
    df_fifa[height_col] = pd.to_numeric(
        df_fifa[height_col].apply(lambda x: extract_metric_value(x, 'cm')),
        errors='coerce',
    )
if weight_col:
    df_fifa[weight_col] = pd.to_numeric(
        df_fifa[weight_col].apply(lambda x: extract_metric_value(x, 'kg')),
        errors='coerce',
    )

fifa_cols = ['nome_join']
if height_col:
    fifa_cols.append(height_col)
if weight_col:
    fifa_cols.append(weight_col)
if foot_col:
    fifa_cols.append(foot_col)
if pos_col:
    fifa_cols.append(pos_col)
df_fifa_reduced = df_fifa[fifa_cols].drop_duplicates(subset=['nome_join'])

# ==========================================
# 5. MERGE
# ==========================================
print('\nMerging datasets (Data Fusion)...')
df_finale = pd.merge(df_fbref, df_fifa_reduced, on='nome_join', how='left')
df_finale = df_finale.drop(columns=['nome_join'])

name_col_final = fbref_name_col if fbref_name_col in df_finale.columns else 'player'
if name_col_final in df_finale.columns:
    bad_mask = df_finale[name_col_final].astype(str).str.contains(MOJIBAKE_PATTERN, na=False)
    bad_count = int(bad_mask.sum())
    print(f'Remaining mojibake in player names: {bad_count}')
    if bad_count > 0:
        print('Residual examples:')
        print(df_finale.loc[bad_mask, name_col_final].head(10).to_string(index=False))

# ==========================================
# 6. REPORT AND SAVE
# ==========================================
if height_col:
    missing_matches = int(df_finale[height_col].isna().sum())
    missing_pct = (missing_matches / len(df_finale)) * 100
    print('\n--- MERGE RESULT ---')
    print(f'Total FBref players: {len(df_finale)}')
    print(f'Players without physical match in EAFC26: {missing_matches} ({missing_pct:.1f}%)')
    print(f'Height in cm (numeric): {height_col}')
    if missing_pct > 30:
        print('WARNING: high missing ratio, name mismatches may exist.')
else:
    print('WARNING: height column not found in EAFC file.')

if weight_col:
    print(f'Weight in kg (numeric): {weight_col}')

if pos_col:
    print(f'Position: {pos_col}')

file_output = 'merged_data.csv'
try:
    df_finale.to_csv(file_output, index=False, encoding='utf-8-sig')
except PermissionError:
    file_output = 'merged_data_new.csv'
    df_finale.to_csv(file_output, index=False, encoding='utf-8-sig')
    print('Original file is locked: saved using an alternative name.')

print(f"\nSave completed: '{file_output}'")
print('🎉 Dataset is ready for machine learning! 🎉')
```
